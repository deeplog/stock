```python

import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 두 개의 독립적인 Figure와 Axes 생성
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

# 첫 번째 그래프
line1, = ax1.plot(x, y1, 'b-', label='sin(x)')
ax1.set_title('Sine Function')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# 두 번째 그래프
line2, = ax2.plot(x, y2, 'r-', label='cos(x)')
ax2.set_title('Cosine Function')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

# 새로운 Figure 생성
fig_combined, ax_combined = plt.subplots(figsize=(10, 6))

# ax1과 ax2의 데이터를 새 Figure에 그리기
ax_combined.plot(line1.get_xdata(), line1.get_ydata(), color=line1.get_color(), label=line1.get_label())
ax_combined.plot(line2.get_xdata(), line2.get_ydata(), color=line2.get_color(), label=line2.get_label())

# x축과 y축 범위 설정
ax_combined.set_xlim(ax1.get_xlim())
ax_combined.set_ylim(min(ax1.get_ylim()[0], ax2.get_ylim()[0]), 
                     max(ax1.get_ylim()[1], ax2.get_ylim()[1]))

# 제목과 레이블 설정
ax_combined.set_title('Overlaid Sine and Cosine Functions')
ax_combined.set_xlabel('x')
ax_combined.set_ylabel('y')

# 범례 추가
ax_combined.legend()

# 그리드 추가 (선택사항)
ax_combined.grid(True, linestyle='--', alpha=0.7)

# 원래의 Figure 닫기 (선택사항)
plt.close(fig1)
plt.close(fig2)

# 새로운 combined Figure 표시
plt.tight_layout()
plt.show()
```
