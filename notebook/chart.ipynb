```python
import yfinance as yf
import mplfinance as mpf
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from datetime import datetime, timedelta
import pandas as pd

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # 윈도우의 경우
# font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"  # 리눅스의 경우
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

def get_stock_data(ticker, period="1mo"):
    """
    야후 파이낸스에서 주식 데이터를 가져오는 함수
    """
    stock = yf.Ticker(ticker)
    end_date = datetime.now()
    start_date = end_date - timedelta(days=30)
    df = stock.history(start=start_date, end=end_date)
    return df

def plot_candlestick_chart(df, title):
    """
    캔들 차트를 그리는 함수
    """
    # 스타일 설정
    mc = mpf.make_marketcolors(up='r', down='b', inherit=True)
    s = mpf.make_mpf_style(marketcolors=mc, rc={'font.family': 'Malgun Gothic'})

    # 거래량 패널 높이 비율 설정
    fig, axlist = mpf.plot(df, type='candle', volume=True, figratio=(16,8), figscale=1.2,
                           title=title, style=s, returnfig=True, 
                           volume_panel=2, panel_ratios=(6,2))

    # x축 날짜 형식 설정
    ax = axlist[0]
    ax.xaxis.set_major_formatter(plt.DateFormatter('%Y-%m-%d'))
    ax.set_xlabel('날짜')
    ax.set_ylabel('가격 (USD)')

    # 거래량 축 레이블 설정
    ax2 = axlist[2]
    ax2.set_ylabel('거래량')

    plt.show()

# 메인 실행 코드

ticker = "NVDA"  # NVIDIA 주식 코드

df = get_stock_data(ticker)

title = f"NVIDIA (NVDA) 최근 한 달 주가 차트\n{df.index[0].strftime('%Y-%m-%d')} ~ {df.index[-1].strftime('%Y-%m-%d')}"
plot_candlestick_chart(df, title)

# 주가 정보 출력
latest_price = df['Close'][-1]
highest_price = df['High'].max()
lowest_price = df['Low'].min()

print(f"NVIDIA (NVDA) 주가 정보 (최근 한 달)")
print(f"최근 종가: ${latest_price:.2f}")
print(f"최고가: ${highest_price:.2f}")
print(f"최저가: ${lowest_price:.2f}")
```
